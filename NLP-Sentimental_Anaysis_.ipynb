```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
def clean_html(text):
    clean = re .compile('<*?>')
    return re.sub(clean, '', text)
```

```python
data['review']=data['review'].apply (clean_html)
```

```python
data['review'].iloc[5]
```

```python
def remove_special(text):
    x =  ' '
    for i in text:
        if i.isalnum():
             x = x+i
        else:
             x = x+ ' '
    return x
```

```python
data['review']=data['review'].apply(remove_special)
```

```python
data['review'].iloc[5]
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
a = stopwords.words ('english')


```

```python
a
```

```python
def remove_stop(text):
    x =[]
    for i in text.split():
        if i not in a :
          x.append(i)
        else:
          pass
    return x
```

```python
data['review']=data['review'].apply(remove_stop)
```

```python
data['review'].iloc[5]
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem_words(text):
    x =[]
    for i in text:
        x.append(ps.stem(i))
    return x
```

```python
data['review']=data['review'].apply(stem_words)
```

```python
data['review'].iloc[5]
```

```python
def join(list_input):
    return ' ',join(list_input)
```

```python
data['review']=data['review'].apply(join)
```
