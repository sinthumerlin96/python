Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=2, random_state=101)
```

```python
data[0][0:5]
```

```python
data[1][0:5]
```

Scaler and Inverse Scaler

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(data[0])
```

```python
trans_data = scaler.transform(data[0])
print(trans_data[0:5])
```

```python
org_data =scaler.inverse_transform(trans_data)
print(org_data[0:5])
```

Visualize Data

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```


Creating the Clusters

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(data[0])
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
kmeans.fit_predict(data[0])
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
import numpy as np
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
```

```python
K = range(1, 10)
meandistortions = []
for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data[0])
    meandistortions.append(sum(np.min(cdist(data[0], kmeans.cluster_centers_, 'euclidean'), axis=1)) / data[0].shape[0])
```

```python
plt.plot(K, meandistortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Average distortion')
plt.title('Selecting k with the Elbow Method')
plt.show()
```

You should note, the colors are meaningless in reference between the two plots.
